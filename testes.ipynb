```python
import pandas as pd

tabela = pd.read_csv("pizzas.csv")
tabela
```

```python
# Plotar um gráfico de dispersão para verificar a correlação
tabela.plot(kind="scatter", x="diametro", y="preco");
```

```python
from sklearn import linear_model

modelo = linear_model.LinearRegression() #criando o modelo

# Para treinar o modelo, precisamos passar os dados de diametro e preço
x = tabela[["diametro"]] #acessando a coluna diametro
y = tabela[["preco"]] #acessando a coluna preço

modelo.fit(x, y)
```

```python
# Agora que o modelo está treinado, podemos fazer previsões
print(modelo.predict([[1]])[0][0]) #prever o preço de uma pizza de determinado cm
```
